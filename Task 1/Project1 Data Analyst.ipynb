```python
#Load the dataset
import pandas as pd
df = pd.read_excel("Dataset for Data Analytics.xlsx")
print(df.head())
```

```python
#Check missing value
print(df.isnull().sum())
```

```python
#Handle missing value with replacing missing value with "no coupon"
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
print(df.isnull().sum())
```

```python
#Check duplicate
print(df.duplicated().sum())
print(df["OrderID"].duplicated().sum())
```

```python
#Verify data format
df["Date"] = pd.to_datetime(df["Date"])
print(df["Date"].dtype)
```

```python
#Save cleaned dataset
df.to_excel("Cleaned Dataset.xlsx", index=False)
```
